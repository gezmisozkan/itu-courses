```python
import matplotlib.pyplot as plt
from scipy import signal

# Define the transfer functions
s1 = signal.lti([1], [10, 10], 1000) # zero at 1, poles at 10, 10 and gain of 1000
s2 = signal.lti([], [10], 10)  # no zero, pole at 10 and gain of 10

# Compute the frequency response
w1, mag1, phase1 = signal.bode(s1)
w2, mag2, phase2 = signal.bode(s2)

# Convert phase to terms of pi
# Normalization and conversion operation on the phase response of a signal so that phase response starts at 0
phase1 = -(phase1 - phase1[0]) / 180
phase2 = -(phase2 - phase2[0]) / 180

# Plot the magnitude response
plt.figure()
plt.semilogx(w1, mag1, label='H1(jw)')  # Bode phase plot system 1
plt.semilogx(w2, mag2, label='H2(jw)')  # Bode phase plot system 2
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Magnitude [dB]')
plt.grid(which='both', axis='both')
plt.legend()
plt.title('Magnitude plots of H1(jw) and H2(jw)')
```

```python
# Plot the phase response
plt.figure()
plt.semilogx(w1, phase1, label='H1(jw)')  # Bode phase plot system 1
plt.semilogx(w2, phase2, label='H2(jw)')  # Bode phase plot system 2
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Phase ['r'$\pi$]')
plt.grid(which='both', axis='both')
plt.legend()
plt.title('Phase plots of H1(jw) and H2(jw)')
```

```python

# Compute the frequency response for the cascade
# Just multiply s1 and s2 to get the cascade
s3 = signal.lti([1], [10, 10, 10], 10000)
w3, mag3, phase3 = signal.bode(s3)

phase3 = -(phase3 - phase3[0]) / 180

# Plot the magnitude and phase response for the cascade
plt.figure()
plt.semilogx(w3, mag3)  # Bode magnitude plot system 1-2
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Magnitude [dB]')
plt.grid(which='both', axis='both')
plt.title('Magnitude plot of H1(jw)H2(jw)')
```

```python

plt.figure()
plt.semilogx(w3, phase3)  # Bode phase plot system 1-2
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Phase ['r'$\pi$]')
plt.grid(which='both', axis='both')
plt.title('Phase plot of H1(jw)H2(jw)')

plt.show()
```

```python

# Plot all the magnitude response
plt.figure()
plt.semilogx(w1, mag1, label='H1(jw)')  # Bode magnitude plot system 1
plt.semilogx(w2, mag2, label='H2(jw)')  # Bode magnitude plot system 2
plt.semilogx(w3, mag3, label='H1(jw)H2(jw)')  # Bode magnitude plot system 1-2
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Magnitude [dB]')
plt.grid(which='both', axis='both')
plt.legend()
plt.title('Bode magnitude plots')

# Plot all the phase response
plt.figure()
plt.semilogx(w1, phase1, label='H1(jw)')  # Bode phase plot system 1
plt.semilogx(w2, phase2, label='H2(jw)')  # Bode phase plot system 2
plt.semilogx(w3, phase3, label='H1(jw)H2(jw)')  # Bode phase plot system 1-2
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Phase ['r'$\pi$]')
plt.grid(which='both', axis='both')
plt.legend()
plt.title('Bode phase plots')
```
